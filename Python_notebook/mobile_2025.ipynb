```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\HP\Downloads\Mobiles Dataset (2025).csv",encoding="latin-1")
```

```python
df = dataset.copy()
```

```python
df.head()
```

```python
df.columns
```

```python
df['Launched Price (Pakistan)']=df['Launched Price (Pakistan)'].str.replace('PKR ', '').str.replace(',', '')
```

```python
df['Launched Price (Pakistan)']=pd.to_numeric(df['Launched Price (Pakistan)'],errors='coerce')
```

```python
df['Launched Price (China)']=df['Launched Price (China)'].str.replace('CNY ',"").str.replace(",","").str.replace( '¥','').replace(' ','')
```

```python
df['Launched Price (China)']=pd.to_numeric(df['Launched Price (China)'],errors='coerce')
df['Launched Price (China)']=df['Launched Price (China)'].astype(float)
```

```python
df['Launched Price (India)']=df['Launched Price (India)'].str.replace('INR ',"").str.replace(",","").str.replace( ' ','')
```

```python
df['Launched Price (India)']=pd.to_numeric(df['Launched Price (India)'],errors='coerce')
df['Launched Price (India)']=df['Launched Price (India)'].astype(float)
```

```python
df['Launched Price (USA)']=df['Launched Price (USA)'].str.replace('USD ',"").str.replace(",","").str.replace( ' ','')
```

```python
df['Launched Price (USA)']=pd.to_numeric(df['Launched Price (USA)'],errors='coerce')
df['Launched Price (USA)']=df['Launched Price (USA)'].astype(float)
```

```python
df['Launched Price (Dubai)']=df['Launched Price (Dubai)'].str.replace('AED ',"").str.replace(",","").str.replace( ' ','')
```

```python
df['Launched Price (Dubai)']=pd.to_numeric(df['Launched Price (Dubai)'],errors='coerce')
df['Launched Price (Dubai)']=df['Launched Price (Dubai)'].astype(float)
```

```python
df['RAM']=df['RAM'].str.replace("GB","")
```

```python
df['RAM']=pd.to_numeric(df['RAM'],errors='coerce')
```

```python
df['Mobile Weight']=df['Mobile Weight'].str.replace('g','')
```

```python
df['Mobile Weight']=pd.to_numeric(df['Mobile Weight'],errors='coerce')
```

```python
df['Front Camera']=df['Front Camera'].str.replace("MP","").str.replace(' (UDC)',"")
```

```python
df['Front Camera']=pd.to_numeric(df['Front Camera'],errors='coerce')
```

```python
df['Back Camera']=df['Back Camera'].str.replace('MP','')
```

```python
df['Back Camera']=pd.to_numeric(df['Back Camera'],errors='coerce')
```

```python
df['Battery Capacity']=df['Battery Capacity'].str.replace('mAh','').str.replace(",","")
```

```python
df['Battery Capacity']=pd.to_numeric(df['Battery Capacity'],errors='coerce')
```

```python
df['Screen Size']=df['Screen Size'].str.replace('inches','')
```

```python
df['Screen Size']=pd.to_numeric(df['Screen Size'],errors='coerce')
```

```python
# Calculate average price by company
avg_price_by_company = df.groupby('Company Name')['Launched Price (India)'].mean().sort_values(ascending=False)

# Plot the average price by company
plt.figure(figsize=(12, 6))
avg_price_by_company.plot(kind='bar')
plt.title('Average Mobile Price by Company in India')
plt.xlabel('Company Name')
plt.ylabel('Average Price (INR)')
plt.show()
```

```python
# Get the top 10 most expensive mobiles
top_10_expensive = df.nlargest(10, 'Launched Price (India)')

# Plot the top 10 most expensive mobiles
plt.figure(figsize=(12, 6))
sns.barplot(x='Model Name', y='Launched Price (India)', data=top_10_expensive)
plt.title('Top 10 Most Expensive Mobiles in Inida')
plt.xlabel('Model Name')
plt.ylabel('Price (INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of screen sizes
plt.figure(figsize=(10, 6))
sns.histplot(df['Screen Size'], bins=20, kde=True)
plt.title('Distribution of Screen Sizes')
plt.xlabel('Screen Size (inches)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by year and calculate the average price
price_trend = df.groupby('Launched Year')['Launched Price (India)'].mean()

# Plot the price trend over the years
plt.figure(figsize=(10, 6))
price_trend.plot(kind='line', marker='o')
plt.title('Price Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price (INR)')
plt.show()
```
