```python
df = dataset.drop(columns=['Restaurant ID','Longitude', 'Latitude','Address','Locality Verbose'])
```

```python
list_of_country = {1:'India',216:'US',215:'UK',30:'Brazil',214:'UAE',189:'South Africa',
                   148:'New Zealand',208:'Turkey',14:'Australia',162:'Philippines',94:'Indonesia',
                   184:'Singapore',166:'Qatar',191:'Sri Lanka',37:'Canada'}

df['Country_name'] = df['Country Code'].map(list_of_country)
```

```python
df.drop(columns=['Country Code'],inplace=True)    
```

```python
df.head()
```

```python
numerical_feature = df.select_dtypes(include=['float','int']).columns.tolist()
category_feature = df.select_dtypes(include='object').columns.tolist()
```

```python
numerical_feature
```

```python
category_feature
```

```python
rows=2
cols=2

fig,ax = plt.subplots(rows,cols,figsize=(20,10))
ax = ax.flatten()

for i,num in enumerate(numerical_feature):
    sns.histplot(data = df, x=num, ax=ax[i], kde=True, bins=30)
    ax[i].set_title(f'Distribution of {num}')
    ax[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
rows=2
cols=2

fig,ax = plt.subplots(rows,cols,figsize=(14,11))
ax = ax.flatten()

for i,num in enumerate(numerical_feature):
    sns.scatterplot(data = df, x=num, y='Aggregate rating',hue='Aggregate rating' ,ax=ax[i])
    ax[i].set_title(f'Distribution of {num}')
    ax[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
y_n = ['Has Table booking',
 'Has Online delivery',
 'Is delivering now',
 'Switch to order menu',
 'Rating color',
 'Rating text']

for i in y_n:
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x=i)
    plt.show()
```
