```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
x = np.linspace(-5, 5, 100)
y1 = -x + 6
y2 = x + 4

fig, ax = plt.subplots(figsize=(12, 7))
ax.scatter(1,5,s=200, zorder=5, color="r", alpha=0.8)

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot([1, 1], [0, 5], ls="--", color="b", alpha=0.5)
ax.plot([-5, 1], [5, 5], ls="--", color="b", alpha=0.5)
ax.set_xlim([-5, 5])
ax.set_ylim([0, 12])

ax.legend()
s = "(1,5)"
ax.text(0.7, 5.6, s, fontsize=20)
ax.grid()
```
